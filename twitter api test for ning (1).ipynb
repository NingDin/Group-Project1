```python

def extract_timeline_as_df(timeline_list):
    columns = set()
    allowed_types = [str, int]
    tweets_data = []
    for x in range(len(timeline_list)):
        try:
            for status in timeline_list[x][0]:
                for y in range(99):
                    status_dict = dict(status)
                    keys = status_dict.keys()
                    single_tweet_data = {"username": timeline_list[x][1]['users'][0]["username"], "date": timeline_list[x].data[y].data}
                    for k in keys:
                        try:
                            v_type = type(status_dict[k])
                        except:
                            v_type = None
                        if v_type != None:
                            if v_type in allowed_types:
                                single_tweet_data[k] = status_dict[k]
                                columns.add(k)
                    tweets_data.append(single_tweet_data)
        except:
            print("nothing found")

    header_cols = list(columns)
    header_cols.append("username")
    header_cols.append('author')
    header_cols.append('date')
    df = pd.DataFrame(tweets_data, columns=header_cols)


    header_cols = list(columns)
    header_cols.append("username")

    df = pd.DataFrame(tweets_data, columns=header_cols)
    return df
```
